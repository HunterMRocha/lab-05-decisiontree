```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("forest_cover.csv")
df = df.sample(2000)
df
```

```python
from sklearn.model_selection import train_test_split

test_df_y = df["Cover_Type"]
del df["Cover_Type"]
test_df_x = df

Y_values = test_df_y.values
X_values = test_df_x.values

x_train, x_test, y_train, y_test = train_test_split(X_values, Y_values, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
lg = LogisticRegression(C=0.000000001, solver='liblinear', max_iter=1000)
rfc = RandomForestClassifier(n_estimators=1000, random_state=42)
dtc = DecisionTreeClassifier()

lg.fit(x_train, y_train)
rfc.fit(x_train, y_train)
dtc.fit(x_train, y_train)
```

```python
y_predLG = lg.predict(x_test)
y_predRFC = rfc.predict(x_test)
y_predDTC = dtc.predict(x_test)

# Calculate Model Accuracy
print("Accuracy:", accuracy_score(y_test, y_predLG))
print("Accuracy:", accuracy_score(y_test, y_predRFC))
print("Accuracy:", accuracy_score(y_test, y_predDTC))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(25,20))
rand_forest_estimators = rfc.estimators_
tree.plot_tree(rand_forest_estimators[0],
               filled = True);
```

#### Samples refers to the number of samples observed in the nodes above. This will allow you to go down the tree since the branches are being split at each step of the class. The valus refer to the class distribution of the samples in the node. The conditions refers to the other features present that could help determine future classifications.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predLG)
cm1 = confusion_matrix(y_test, y_predDTC)
cm2 = confusion_matrix(y_test, y_predRFC)

print("random forest cm", cm2)
print('decision tree cm', cm1)
print('logistic regression cm', cm)
```

Since there are a lot of features, random forest is the most accurate because the decisions are made based on the subset of the features. The permutations of feature combination allow for the model to make better predictive decisions
